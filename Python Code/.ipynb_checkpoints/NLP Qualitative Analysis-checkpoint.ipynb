## Qualitative Analysis - taking NewsAPI articles on the Canadian Housing Market 

Use the [newsapi](https://newsapi.org/) to pull the latest news articles for the Canadian Housing Market

In this part of our project, we apply natural language processing inorder to understand the sentiment in the latest news articles featuring the Canadian Housing Market. We also try to use the fundamental NLP techniques to try and understand the other factors involved with the housing market such as common words and phrases and organizations and entities mentioned in the articles.

Our analysis spans out as follows:

1. Sentiment Analysis - We use the newsapi to pull the latest news articles for teh Canadian Housing Market and create a DataFrame of sentiment scores.
2. Natural Language Processing - We use NLTK and Python to tokenize text, find n-gram counts, and create a word cloud.
3. Named Entity Recognition - We try to build a named entity recognition model for the Canadian Housing Market and visualize the tags using SpaCy.


```python
# Initial imports
import json
import nltk as nltk
import os
import pandas as pd
import requests
import spacy
from dotenv import load_dotenv
from newsapi import NewsApiClient
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import word_tokenize, sent_tokenize
from path import Path
from spacy import displacy
nltk.download('vader_lexicon')
from string import punctuation
import re
from collections import Counter
from nltk import ngrams

%matplotlib inline
```

```python
# Read your api key environment variable
# Load environment variables and retrieve the News API key
load_dotenv()
api_key = os.getenv("NEWS_API_KEY")
```

```python
# Create the newsapi client
newsapi = NewsApiClient(api_key=api_key)
natural_processing = spacy.load('en_core_web_sm')
```

```python
# Fetch Canadian Housing Market
housingmarket_news_en = newsapi.get_everything(q="(Canadian OR Canada) AND Housing Market",
                                               language="en",
                                               sort_by='relevancy'
)

# Show the total number of news
print(f"Total articles about Housing Market: {housingmarket_news_en['totalResults']}")
```

```python
# Create the Canadian Housing Market sentiment scores DataFrame
canhm_sentiments = []
for article in housingmarket_news_en["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        # append each element of each iteration to the btc_sentiments empty list
        
        canhm_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
        })
    except AttributeError:
        pass
    
canhm_sentiments_df = pd.DataFrame(canhm_sentiments)
cols = ["date", "text", "compound", "positive", "negative", "neutral"]
canhm_sentiments_df = canhm_sentiments_df[cols]

canhm_sentiments_df.head()
```

```python
# Describe sentiment analysis
canhm_sentiments_df.describe()
```

### Tokenizer

```python
# Instantiate the lemmatizer
lemmatizer = WordNetLemmatizer()

# Create a list of stopwords
sw = set(stopwords.words('english'))

# Expand the default stopwords list if necessary
stopwords_addons = {'said', 'sent', 'found', 'canadian', 'reuters', 'market','including','new', 'picture', 'today', 'announced', 'week', 'basically', 'also', 'chars', '\r\n'}
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    
    # Remove the punctuation from text
    
    words = nltk.word_tokenize(text)
    words = [word.lower() for word in words if word.isalpha()]
    str_words = ' '.join(str(e) for e in words)
    
    # Create a tokenized list of the words
    
    doc = natural_processing(str_words)
    doc_words = ' '.join(str(e) for e in doc)
    tok = word_tokenize(doc_words)
    result = [word.lower() for word in tok]
    lemma_words = ' '.join(str(e) for e in result)
    
    # Lemmatize words into root words
    
    lemma_result = lemmatizer.lemmatize(lemma_words)
   
    # Convert the words to lowercase
    
    lower_result = [lemma_result.lower() for word in lemma_result]
    re_words = ' '.join(str(e) for e in lower_result)
    
    # Remove the stop words
    
    re_words = word_tokenize(re_words)
    tokens = [word.lower() for word in re_words if word.lower() not in sw.union(stopwords_addons)]
    
    
    return tokens
```

```python
# Check the datatypes of the sentiment analysis dataframe
canhm_sentiments_df.dtypes
```

```python
# Convert date from object to datetime format
canhm_sentiments_df['date'] = canhm_sentiments_df['date'].astype('datetime64[ns]')
canhm_sentiments_df.info()
```

```python
# Set the date as the index and sort
canhm_sentiments_df = canhm_sentiments_df.set_index('date')
canhm_sentiments_df = canhm_sentiments_df.sort_index()
```

```python
# Create a new tokens column for Canadian Housing Market
canhm_sentiments_df["tokens"] = canhm_sentiments_df["text"].apply(tokenizer)
canhm_sentiments_df.head()
```

```python
# Plot the sentiment analysis
canhm_sentiments_df.plot(kind='bar', figsize=(15,7))
```

```python
# Plot the sentiment analysis for positive and negative only
canhm_sentiments_df_pn = canhm_sentiments_df.drop(canhm_sentiments_df.columns.difference(['positive', 'negative']), 1)
canhm_sentiments_df_pn.plot(kind='bar', figsize=(15,7))
```

```python
# Generate the Canadian Housing Market N-grams where N=2
canhm_grams = ngrams(tokenizer(canhm_sentiments_df["text"].str.cat()), n=2)
Counter(canhm_grams).most_common(20)
```

```python
# Function token_count generates the top 10 words for a subject
def token_count(tokens, N=3):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Use token_count to get the top 10 words for Canadian Housing Market
canhm_all = tokenizer(canhm_sentiments_df["text"].str.cat())
result_top25_canhm = token_count(canhm_all, 25)
print(result_top25_canhm)
```

```python
# Generate the Canadian Housing Market word cloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [25.0, 12.0]
```

```python
# Generate the Canadian Housing Market word cloud
terms_list_canhm = str(result_top25_canhm)

# Create the word cloud
wordcloud = WordCloud(width = 1000, height = 500,background_color ='white',min_font_size = 10).generate(terms_list_canhm)
plt.imshow(wordcloud)
plt.axis("off")
fontdict = {"fontsize": 25, "fontweight": "bold"}
plt.title("Canadian Housing Market", fontdict=fontdict)
plt.imshow(wordcloud)
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```
